```python
# Further analysis

df_grouped = df.groupby('Mapped Beach').mean()
```

```python
rules_map = {
    'level_of_interaction_with_disaster': 1.50,
    'frequency_of_usage': 5.50,
    'usage_duration': 17.50,
    'number_of_lik_combination': 1.50,
    'number_of_experience_with_disaster': 1.50,
}

df_grouped['Level of Interaction with Disaster Status'] = np.where(df_grouped['Level of Interaction with Disaster'] > 1.50, 'Safe', 'Unsafe')
df_grouped['Frequency of Usage Status'] = np.where(df_grouped['Frequency of Usage (max) (in month)'] > 5.50, 'Safe', 'Unsafe')
df_grouped['Usage Duration Status'] = np.where(df_grouped['Usage Duration'] > 17.50, 'Safe', 'Unsafe')
df_grouped['Number of LIK Combination Status'] = np.where(df_grouped['Number of LIK Combination'] > 1.50, 'Safe', 'Unsafe')
df_grouped['Number of Experience with Disaster Status'] = np.where(df_grouped['Number of Experience with Disaster'] > 1.50, 'Safe', 'Unsafe'))
```

```python
df_grouped.head()
```

```python
df_grouped['count'] = (df_grouped == 'Unsafe').sum(axis=1)
max_count = df_grouped['count'].max()
```

```python
df_grouped['Overall Category'] = np.where(df_grouped['count'] >= max_count, 'Unsafe', 'Safe')
df_grouped.drop(columns=['count'], inplace=True)

df_grouped
```

## Recommendation Action Analysis

```python
df = pd.read_csv('data/processed-recommendation.csv')
df.head()
```

```python
action_column = 'Action Anticipation (Respon based on LIK codes)' 

df['Cleaned_Action'] = df[action_column].astype(str).str.split(':').str[-1].str.strip()

lik_columns = [col for col in df.columns if str(col).startswith('Wn-')]

results = []

for lik in lik_columns:
    df_filtered = df[df[lik] == 1]
    
    if not df_filtered.empty:
        modes = df_filtered['Cleaned_Action'].mode()
        
        if not modes.empty:
            most_action = modes.iloc[0]
        else:
            most_action = "Tidak ada aksi"
    else:
        most_action = "Tidak ada kemunculan LIK"
        
    results.append({
        'LIK': lik,
        'most_action_taken': most_action
    })

df_result = pd.DataFrame(results)

print("--- Most Action Taken per LIK ---")
display(df_result)
```

```python
target_wn = ['Wn-2', 'Wn-8', 'Wn-1', 'Wn-3', 'Wn-7', 'Wn-4', 'Wn-9', 'Wn-13', 'Wn-6', 'Wn-5']

df_filtered_lik = df_result[df_result['LIK'].isin(target_wn)].copy()

df_filtered_lik['Wn_Num'] = df_filtered_lik['LIK'].str.replace('Wn-', '').astype(int)

df_filtered_lik = df_filtered_lik.sort_values('Wn_Num', ascending=True)

df_filtered_lik = df_filtered_lik.drop(columns=['Wn_Num']).reset_index(drop=True)

action_translation_map = {
    'Stay alert/retrieve fishing gear': 'Siaga Penuh / Amankan Alat Tangkap',
    'Be careful': 'Berhati-hati / Tingkatkan kewaspadaan',
    'Adjust fishing time': 'Sesuaikan Jadwal Melaut'
}

df_filtered_lik['most_action_taken'] = df_filtered_lik['most_action_taken'].map(action_translation_map)
```

```python
df_filtered_lik.to_csv('lik_filtered_action_taken.csv', index=False)

print("--- Filtered & Sorted Action Taken per LIK ---")
display(df_filtered_lik)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/data-master.csv', sep=';')
df.shape
```

```python
selected_cols_to_analyze = ['Level of Interaction with Disaster', 'Frequency of Usage (max) (in month)', 'Usage Duration', 
                            'Number of LIK Combination', 'Number of Experience with Disaster', 'Age', 'Fishing Experience', 'Beach']

df = df[selected_cols_to_analyze]
df.columns
```

```python
beach_map = {
    1: 'pantai lampuuk',
    2: 'pantai lhoknga',
    3: 'pantai ulee lheue',
    4: 'pantai depok',
    5: 'pantai samas',
}

df['Mapped Beach'] = df['Beach'].map(beach_map)

df.drop(columns=['Beach'], inplace=True)

print('Data cleaned')
```

```python
df_pantai_lampuuk = df[df['Mapped Beach'] == 'pantai lampuuk']
df_pantai_lhoknga = df[df['Mapped Beach'] == 'pantai lhoknga']
df_pantai_ulee_lheue = df[df['Mapped Beach'] == 'pantai ulee lheue']
df_pantai_depok = df[df['Mapped Beach'] == 'pantai depok']
df_pantai_samas = df[df['Mapped Beach'] == 'pantai samas']
```

```python
# Further analysis

df_grouped = df.groupby('Mapped Beach').mean()
```

```python
rules_map = {
    'level_of_interaction_with_disaster': 1.50,
    'frequency_of_usage': 5.50,
    'usage_duration': 17.50,
    'number_of_lik_combination': 1.50,
    'number_of_experience_with_disaster': 1.50,
}

df_grouped['Level of Interaction with Disaster Status'] = np.where(df_grouped['Level of Interaction with Disaster'] > 1.50, 'Safe', 'Unsafe')
df_grouped['Frequency of Usage Status'] = np.where(df_grouped['Frequency of Usage (max) (in month)'] > 5.50, 'Safe', 'Unsafe')
df_grouped['Usage Duration Status'] = np.where(df_grouped['Usage Duration'] > 17.50, 'Safe', 'Unsafe')
df_grouped['Number of LIK Combination Status'] = np.where(df_grouped['Number of LIK Combination'] > 1.50, 'Safe', 'Unsafe')
df_grouped['Number of Experience with Disaster Status'] = np.where(df_grouped['Number of Experience with Disaster'] > 1.50, 'Safe', 'Unsafe'))
```

```python
df_grouped.head()
```

```python
df_grouped['count'] = (df_grouped == 'Unsafe').sum(axis=1)
max_count = df_grouped['count'].max()
```

```python
df_grouped['Overall Category'] = np.where(df_grouped['count'] >= max_count, 'Unsafe', 'Safe')
df_grouped.drop(columns=['count'], inplace=True)

df_grouped
```

```python
# Save the rule matrix
df_grouped.to_csv('data/community_rule_categorization.csv')
```
